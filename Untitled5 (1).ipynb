```python
fs = 1000.0  # Hz

def cut_segment_all(time, emg, t_start, t_end, fs):
    i0 = int(round(t_start * fs))
    i1 = int(round(t_end   * fs))
    i0 = max(0, i0)
    i1 = min(len(time), i1)
    return time[i0:i1], emg[i0:i1, :]   # all 4 channels

segments = []
for idx, (t_start, t_end) in enumerate(boundaries, start=1):
    seg_time, seg_emg = cut_segment_all(time, emg, t_start, t_end, fs)
    print(f"Trial {idx}: {seg_time[0]:.2f}s to {seg_time[-1]:.2f}s, samples = {len(seg_time)}")
    segments.append((seg_time, seg_emg))
```

```python
import numpy as np
from scipy.signal import butter, filtfilt, iirnotch

def bandpass_filter(x, fs, low=10, high=450, order=4):
    nyq = 0.5 * fs
    b, a = butter(order, [low/nyq, high/nyq], btype='band')
    return filtfilt(b, a, x, axis=0)

def notch_50hz(x, fs, f0=50.0, Q=30.0):
    nyq = 0.5 * fs
    b, a = iirnotch(f0/nyq, Q)
    return filtfilt(b, a, x, axis=0)

def rectify_and_envelope(x, fs, lp_cut=10, order=4):
    rect = np.abs(x)
    nyq = 0.5 * fs
    b, a = butter(order, lp_cut/nyq, btype='low')
    env = filtfilt(b, a, rect, axis=0)
    return rect, env
```

```python
preprocessed_segments = []

for idx, (seg_time, seg_emg) in enumerate(segments, start=1):
    # 1) bandâpass
    emg_bp = bandpass_filter(seg_emg, fs)
    # 2) notch
    emg_bp_notch = notch_50hz(emg_bp, fs)
    # 3) rectification + envelope
    rect_seg, env_seg = rectify_and_envelope(emg_bp_notch, fs)

    preprocessed_segments.append({
        "time": seg_time,
        "raw": seg_emg,
        "bp_notch": emg_bp_notch,
        "rect": rect_seg,
        "env": env_seg
    })

    print(f"Trial {idx}: preprocessed (shape {emg_bp_notch.shape})")
```

```python
import pandas as pd

for idx, seg in enumerate(preprocessed_segments, start=1):
    seg_time = seg["time"]
    emg_proc = seg["bp_notch"]   # or seg["env"] if you want envelopes

    out = np.column_stack([seg_time, emg_proc])
    cols = ['time'] + [f'ch{i+1}' for i in range(emg_proc.shape[1])]
    df_seg = pd.DataFrame(out, columns=cols)

    out_name = f'{filename}_trial{idx}_bp_notch.csv'
    df_seg.to_csv(out_name, index=False)
    print("Saved:", out_name)
```
